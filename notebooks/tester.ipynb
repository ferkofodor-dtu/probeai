```python
from data.make_dataset import load_probeai
from pathlib import Path

path = Path.cwd().parents[0] / 'data' / 'raw'
train, test = load_probeai(src=path)
```

```python
import torch
from torch.utils.data import DataLoader
from pathlib import Path
from data.make_dataset import ProbeAIDataset

path_p = Path.cwd().parents[0] / 'data' / 'processed'

train_processed = torch.load(path_p / 'train_data.pt')
test_processed = torch.load(path_p / 'test_data.pt')
```

```python
batch_size = 8

train_loader = DataLoader(train, batch_size=batch_size, shuffle=False)
test_loader = DataLoader(test, batch_size=batch_size, shuffle=False)

tl = DataLoader(train_processed, batch_size=8, shuffle=False)
tt = DataLoader(test_processed, batch_size=8, shuffle=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import torchvision

# functions to show an image
def imshow(img):
    img = img / 2 + 0.5     # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))
    plt.show()


# get some random training images
dataiter = iter(train_loader)
test_iter = iter(test_loader)
```

```python
images, labels = next(iter(tl))

# show images
imshow(torchvision.utils.make_grid(images))
# print labels
print('Train GroundTruth '+' '.join(f'{labels[j]}' for j in range(batch_size)))

images, labels = next(iter(tt))

# show images
imshow(torchvision.utils.make_grid(images))

# print labels
print('Test GroundTruth '+' '.join(f'{labels[j]}' for j in range(batch_size)))
```

```python
images2, labels = next(dataiter)

# show images
imshow(torchvision.utils.make_grid(images2))
# print labels
print('Train GroundTruth '+' '.join(f'{labels[j]}' for j in range(batch_size)))

images2, labels = next(test_iter)

# show images
imshow(torchvision.utils.make_grid(images2))

# print labels
print('Test GroundTruth '+' '.join(f'{labels[j]}' for j in range(batch_size)))
```
